```python
import seaborn as sns
```

```python
def download_file(link):
    import wget
    from os.path import exists
    file_name = link.split('/')[-1]
    if not exists('data/'+file_name):
        wget.download(link,'data')
```

```python
download_file('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
```

```python
df = sns.load_dataset('diamonds')
```

```python
sns.scatterplot(data=df,x='carat',y='price', hue='clarity', size='depth', style='cut')
```

```python
sns.catplot(data=df,x='cut',y='price',kind='bar')
```
