```python
import h2o
from h2o.automl import H2OAutoML
import wget
from os.path import exists
```

```python
h2o.init(nthreads = -1, max_mem_size = 16)
```

```python
if not exists('data/ml/wine.data'):
    wget.download('https://raw.githubusercontent.com/vsvale/datasets/main/wine.data','data')
```

```python
df = h2o.import_file('data/wine.data')
df
```

```python
y = 'C1'
x = list(df.columns).remove(y)
```

```python
df[y] = df[y].asfactor()
```

```python
splits = df.split_frame(ratios=[0.7, 0.15], seed=1)
```

```python
train = splits[0]
valid = splits[1]
test = splits[2]
```

```python
print(train.nrow)
print(valid.nrow)
print(test.nrow)
```

```python
aml = H2OAutoML()
```

```python
aml.train(x = x, y = y, training_frame = train)
```

```python
aml.leaderboard
```

```python
perf = aml.leader.model_performance(test)
```

```python
perf.confusion_matrix
```

```python
aml.leader
```

```python
preds = aml.leader.predict(valid)
preds
```
