```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
mes = ['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN', 'JUL', 'AGO', 'SET', 'OUT','NOV','DEZ'] 
temperatura = [27,27,27,25,23,21,21,22,22,23,25,26]
```

```python
plt.figure(figsize=(6,3))
plt.plot(mes,temperatura, c='green')
plt.xlabel('Month') 
plt.ylabel('Temperature')
plt.ylim(20,30)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(mes,temperatura)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(mes,temperatura)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.scatter(mes,temperatura)
plt.show()
```

```python
values = [15,30,45,10]
labels = ['Frogs','Hogs','Dogs','Logs',]
explode = [0,.1,0,0]
colors=['tab:blue','tab:orange','tab:green','tab:red']
```

```python
plt.figure(figsize=(8,5))
plt.pie(values,labels=labels,explode=explode,colors=colors, autopct='%.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
x = [.8,2.4,3.3,4,4.4,5]
y= [63,34,22,48,18,5]
xticks = np.arange(1.5,5.1,0.5)
yticks = np.arange(20,70,10)
size = [200,120,60,160,40,200]
color = ['red','green','yellow','yellow','green','red']
```

```python
plt.figure(figsize=(7,4))
plt.scatter(x,y,s=size,c=color)
plt.xticks(xticks)
plt.yticks(yticks)
plt.show()
```

```python
x = ['Python','C++','Java','Perl','Scala','Lisp']
y = [10,8,6,4,2,1]
```

```python
plt.figure(figsize=(6,3))
plt.bar(x,y,align='center',alpha=0.5, color='mediumslateblue',edgecolor='black',linewidth=.6)
plt.title('Programming language usage') 
plt.ylabel('Usage')
plt.ylim(0,10)
plt.show()
```
