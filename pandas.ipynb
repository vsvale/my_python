# Vendas

Desafio:

Você faz parte da equipe de Analytics de uma grande marca de vestuário com mais de 25 lojas espalhadas em Shoppings de todo o Brasil.

Toda semana você precisa enviar para a diretoria um ranking atualizado com as 25 lojas contendo 3 informações:

    Faturamento de cada Loja
    Quantidade de Produtos Vendidos de cada Loja
    Ticket Médio dos Produto de cada Loja

Além disso, cada loja tem 1 gerente que precisa receber o resumo das informações da loja dele. Por isso, cada gerente deve receber no e-mail:

    Faturamento da sua loja
    Quantidade de Produtos Vendidos da sua loja
    Ticket Médio dos Produto da sua Loja

Esse relatório é sempre enviado como um resumo de todos os dados disponíveis no ano.


```python
import pandas as pd
import sqlite3
import wget
from os.path import exists
from unidecode import unidecode
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
if not exists('data/Vendas.csv'):
    wget.download('https://raw.githubusercontent.com/vsvale/datasets/main/Vendas.csv','data')
```

```python
df = pd.read_csv('data/Vendas.csv')
df.head()
```

```python
con = sqlite3.connect('VendasDB.db')
```

```python
cur = con.cursor()
```

```python
table_fields = list(df.columns.ravel())
table_fields
```

```python
table_fields = [unidecode(column).replace(" ","_") for column in table_fields]
table_fields
```

```python
cur.execute("""CREATE TABLE IF NOT EXISTS """+"""VENDAS"""+"""("""+','.join(table_fields)+""")""")
```

```python
query = """INSERT INTO"""+""" VENDAS"""+"""("""+','.join(table_fields)+""") VALUES ("""+','.join(map(str,'?'*len(df.columns)))+""")"""
query
```

```python
str_dataset = df.astype(str)
```

```python
for i in range(len(str_dataset)):
  insert = tuple(str_dataset.iloc[i])
  cur.execute(query,insert)
con.commit()
```

```python
df = pd.read_sql_query('SELECT * FROM VENDAS',con)
```

```python
con.close()
```

```python
df['Quantidade'] = pd.to_numeric(df['Quantidade'])
df[['Valor_Unitario','Valor_Final']] = df[['Valor_Unitario','Valor_Final']].astype(float)
df['Data'] = pd.to_datetime(df['Data'])
```

### Calcular o faturamento por Loja

```python
loja_vf = df[['ID_Loja','Valor_Final']]
fat_loja = loja_vf.groupby('ID_Loja').sum()
fat_loja = fat_loja.sort_values(by='Valor_Final',ascending=False)
fat_loja
```

### Calcular Quantidade de produtos vendidos por loja

```python
qtd_loja = df[['Quantidade','ID_Loja']].groupby('ID_Loja').sum().sort_values(by='Quantidade',ascending=False)
qtd_loja
```

### Calcular o ticket Médio dos Produtos por Loja

```python
ticket_medio = (fat_loja['Valor_Final'] /qtd_loja['Quantidade']).to_frame().rename(columns={0:'Ticket_Medio'}).sort_values(by='Ticket_Medio',ascending=False)
ticket_medio
```

# Prova Redação

Crie um data frame com 1000 amostras em cada uma das seguintes colunas:
- Idade: int entre 0 e 100
- Data: date entre 01/01/2000 e 31/01/2000
- Nota: float entre 0 e 1000
- Sexo: M ou F
- Estado

```python
from random import uniform,choices
import numpy as np
import pandas as pd
import seaborn as sns
from sqlalchemy import create_engine
import sqlite3
```

```python
states = {
    'AC': 'Acre',
    'AL': 'Alagoas',
    'AP': 'Amapá',
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Ceará',
    'DF': 'Distrito Federal',
    'ES': 'Espírito Santo',
    'GO': 'Goiás',
    'MA': 'Maranhão',
    'MT': 'Mato Grosso',
    'MS': 'Mato Grosso do Sul',
    'MG': 'Minas Gerais',
    'PA': 'Pará',
    'PB': 'Paraíba',
    'PR': 'Paraná',
    'PE': 'Pernambuco',
    'PI': 'Piauí',
    'RJ': 'Rio de Janeiro',
    'RN': 'Rio Grande do Norte',
    'RS': 'Rio Grande do Sul',
    'RO': 'Rondônia',
    'RR': 'Roraima',
    'SC': 'Santa Catarina',
    'SP': 'São Paulo',
    'SE': 'Sergipe',
    'TO': 'Tocantins'
}
```

```python
idade = choices(range(101), k=1000)
data = choices(pd.date_range('2000-01-01','2000-01-31'),k=1000)
nota = [uniform(0, 1000) for p in range(1000)]
sexo = choices(['M','F'], k=1000)
estado = choices([*states.values()], k=1000)
```

```python
df = pd.DataFrame({
    'idade':idade,
    'data':data,
    'nota':nota,
    'sexo':sexo,
    'estado':estado
})
df
```

Transforme 20% das notas em valores nulos, simulando alunos que não compareceram a prova

```python
df.loc[df.sample(frac=.2,random_state=0).index,'nota'] = np.nan
```

Preencha as notas nulos com valor 0

```python
df.nota.fillna(0,inplace=True)
```

remova alunos com idades inferiores a 18 e superiores a 80

```python
df = df.loc[(df['idade'] >= 18) & (df['idade'] <= 80)]
```

crie um novo campo de aprovado para os alunos com nota superior a 600

```python
df['aprovado'] = df['nota'].apply(lambda x: True if x>= 600 else False)
df
```

crie um campo novo indicando o dia da semana para todas as datas

```python
df['diasemana'] = df['data'].dt.day_of_week
```

crie uma tabela de cruzando sexo por estado

```python
pd.crosstab(df['estado'],df['sexo'])
```

grafico de pizza da quantidade de aprovados por sexo

```python
df.loc[df['aprovado'] == True,'sexo'].value_counts().plot(kind='pie');
```

grafico de barras com a quantidade de aprovados por estado

```python
df.loc[df['aprovado'] == True,'estado'].value_counts().plot(kind='bar');
```

grafico de ponto de nota por idade, colorindo por sexo

```python
sns.scatterplot(
    data=df[df['aprovado'] == True],
    x='idade',
    y='nota',
    hue='sexo'
);
```

grafico de barras com a participação por dia da semana

```python
mapa = {
    0: 'Seg',
    1: 'Ter',
    2: 'Qua',
    3: 'Qui',
    4: 'Sex',
    5: 'Sab',
    6: 'Dom'
}
df['diasemana'].map(mapa).value_counts().plot(kind='bar');
```

grafico de barras com a participação por dia do mes

```python
df['data'].dt.day.value_counts().plot(kind='bar');
```

grafico de pontos das notas por dia da semana

```python
df.groupby('diasemana')['nota'].mean().rename(mapa).plot(kind='bar');
```

salve em csv as notas dos 100 melhores alunos ordenados por nota decrescente

```python
df.sort_values('nota',ascending=False).iloc[:100,:-2].to_csv('data/convocados.csv',index=False)
```

```python
engine = create_engine('sqlite:///db.sqlite3')
```
